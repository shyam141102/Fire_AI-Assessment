```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Discount Name'].fillna('No Discount', inplace=True)
```

```python
df.isnull().sum()
```

```python
numerical_columns = ['Discount Amount (INR)', 'Gross Amount', 'Net Amount']

mean_values = df[numerical_columns].mean()
median_values = df[numerical_columns].median()
mode_values = df[numerical_columns].mode().iloc[0]
std_values = df[numerical_columns].std()

print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)
print("\nStandard Deviation Values:\n", std_values)
```

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_df.corr()
print(correlation_matrix)
```

```python
# Create a heatmap to visualize correlations

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Box plot for Net Amount by Product Category

plt.figure(figsize=(10, 6))
sns.boxplot(x='Product Category', y='Net Amount', data=df)
plt.title('Net Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Net Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for Gross Amount by Discount Availed

plt.figure(figsize=(10, 6))
sns.boxplot(x='Discount Availed', y='Gross Amount', data=df)
plt.title('Gross Amount by Discount Availed Status')
plt.xlabel('Discount Availed')
plt.ylabel('Gross Amount')
plt.show()
```

```python
# Box plot for Net Amount by Age Group

plt.figure(figsize=(10, 6))
sns.boxplot(x='Age Group', y='Net Amount', data=df)
plt.title('Net Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Net Amount')
plt.show()
```

```python
# Bar plot showing the count of purchases per Product Category

plt.figure(figsize=(10, 6))
sns.countplot(x='Product Category', data=df, palette='viridis')
plt.title('Total Purchases per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot showing the count of purchases by Gender

plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='coolwarm')
plt.title('Total Purchases by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
# Bar plot showing the count of Discount Availed per Age Group

plt.figure(figsize=(10, 6))
sns.barplot(x='Age Group', y='Discount Amount (INR)', data=df, estimator=sum, palette='magma')
plt.title('Total Discount Availed by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Discount Amount (INR)')
plt.show()
```

```python
# Bar plot showing count of each Purchase Method

plt.figure(figsize=(14, 6))
sns.countplot(x='Purchase Method', data=df, palette='Set2')
plt.title('Usage of Different Purchase Methods')
plt.xlabel('Purchase Method')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for Gross Amount per Location

plt.figure(figsize=(12, 6))
sns.barplot(x='Location', y='Gross Amount', data=df, estimator=sum, palette='Set3')
plt.title('Total Gross Amount by Location')
plt.xlabel('Location')
plt.ylabel('Gross Amount (INR)')
plt.xticks(rotation=45)
plt.show()
```
